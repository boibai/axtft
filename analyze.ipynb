## 03. Call LLM

```python
import httpx
from app.langgraph.common.schema import ErrorCauseList
```

```python
# 비동기 HTTP 클라이언트 전역 생성 ( 재사용을 통해 커넥션 풀 활용 및 성능 최적화 )
client = httpx.AsyncClient(
    timeout=httpx.Timeout(
        connect=5.0,   # TCP 연결 수립 최대 대기 시간
        read=120.0,    # 응답 바디 수신 최대 대기 시간 (LLM 추론 고려)
        write=30.0,    # 요청 바디 전송 최대 시간
        pool=5.0       # 커넥션 풀에서 대기 가능한 최대 시간
    ),
    limits=httpx.Limits(
        max_connections=50,          # 동시에 열 수 있는 최대 TCP 연결 수
        max_keepalive_connections=20 # Keep-Alive 유지 연결 수
    ),
)
```

```python
MODEL_NAME = "gpt-oss-20b"
VLLM_BASE_URL = "http://10.122.100.172:8000/v1/chat/completions"
```

```python
payload = {
    "model": MODEL_NAME,
    # 사용할 LLM 모델 이름 (예: gpt-oss-20b)

    "messages": state["messages"],
    # LLM 입력 메시지 (system / user / assistant 형태의 대화 히스토리)

    "temperature": 0.0,
    # 생성 다양성 제어 (0.0 = 가장 deterministic, 항상 같은 결과에 가깝게)

    "max_tokens": 4096,
    # LLM이 생성할 최대 토큰 수 (응답 길이 제한)

    "response_format": {
        "type": "json_schema",
        # 응답을 JSON 형태로 강제 (구조화된 출력)

        "json_schema": {
            "name": "cause-list",
            # 스키마 이름 (식별용)

            "schema": ErrorCauseList.model_json_schema(),
            # Pydantic 모델 기반 JSON Schema
            # → LLM이 반드시 이 구조에 맞게 응답하도록 강제
        },
    },
}
```

```python
resp = await client.post(VLLM_BASE_URL, json=payload)
# vLLM 서버로 비동기 POST 요청 전송 (payload: 모델, 메시지, 옵션 포함)

resp.raise_for_status()
# HTTP 응답 상태 체크 (200이 아니면 예외 발생 → 오류 즉시 탐지)

data = resp.json()
# 응답을 JSON 형태로 변환 (LLM 결과 + 메타데이터 포함)

state["llm_content"] = data["choices"][0]["message"]["content"]
# LLM이 생성한 실제 응답 텍스트 추출 및 state에 저장
# (OpenAI/vLLM 응답 구조: choices[0].message.content)

usage = data.get("usage", {})
# 토큰 사용량 정보 추출 (없을 수도 있으므로 get 사용)

state["model_name"] = MODEL_NAME
# 사용된 모델 이름 기록 (추적/로그용)

state["prompt_tokens"] = usage.get("prompt_tokens")
# 입력(prompt)에 사용된 토큰 수

state["completion_tokens"] = usage.get("completion_tokens")
# 출력(response)에 사용된 토큰 수

state["total_tokens"] = usage.get("total_tokens")
# 전체 토큰 사용량 (비용/성능 모니터링용)
```

```python
state
```

## 04. Parse Json

```python
try:
    state["parsed_json"] = json.loads(state["llm_content"])
    # LLM이 생성한 문자열(JSON 형태)을 실제 Python dict로 파싱

    state["error"] = None
    # 정상적으로 파싱되었으므로 에러 상태 초기화

except Exception as e:

    state["parsed_json"] = None
    # JSON 파싱 실패 시 결과를 None으로 설정

    state["error"] = f"JSON parse failed: {e}"
    # 파싱 실패 원인을 state에 기록 (디버깅 및 후속 처리용)
```

```python
state
```

## 05. Validate Schema

```python
try:
    ErrorCauseList.model_validate(state["parsed_json"])
    # 파싱된 JSON이 ErrorCauseList 스키마(Pydantic 모델)에 맞는지 검증

    state["error"] = None
    # 스키마 검증이 성공하면 에러 상태 초기화

except Exception as e:
    state["error"] = f"Schema validation failed: {e}"
    # 스키마에 맞지 않을 경우 에러 메시지를 기록 (구조 불일치, 필드 누락 등)
```
